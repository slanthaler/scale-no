```python
import torch

x = torch.linspace(0,1,100)
X,Y = torch.meshgrid(x,x)
```

```python
def rotateLeft(a):
    # 
    BC1 = a[0, :]
    coeff1 = a[:,:]

    return coeff1, BC1

def rotateTop(a):
    #
    BC2 = a[:,-1]
    coeff2 = torch.flip(a[:,:].transpose(0,1), dims=[0])

    return coeff2, BC2

def rotateRight(a):    
    #
    BC3 = torch.flip( a[-1, :], dims=[0])
    coeff3 = torch.flip(a[:,:], dims=[0,1])

    return coeff3, BC3

def rotateBottom(a):
    #
    BC4 = torch.flip( a[:,0], dims=[0])
    coeff4 = torch.flip(a[:,:].transpose(0,1), dims=[1])

    return coeff4, BC4


a = Y
L, bcL = rotateLeft(a)
T, bcT = rotateTop(a)
R, bcR = rotateRight(a)
B, bcB = rotateBottom(a)

import matplotlib.pyplot as plt

def plotstuff(a,bc):   
    plt.figure()
    plt.pcolor(X,Y,a)
    plt.colorbar()
    plt.show()

    plt.figure()
    plt.plot(bc)
    plt.show()

plotstuff(L,bcL)
plotstuff(T,bcT)
plotstuff(R,bcR)
plotstuff(B,bcB)
```
